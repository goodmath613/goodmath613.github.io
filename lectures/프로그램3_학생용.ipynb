# 저장하기 & 불러오기

| 활용법 | 설명 |
| :---: | :---:|
|데이터.to_csv('파일명', index=False) | 데이터를 파일명으로 저장한다. 단, 인덱스는 저장하지 않는다. |
|데이터.to_csv('파일명', index=True) | 데이터를 파일명으로 저장한다. 단, 인덱스도 포함하여 저장한다. |
|pd.read_csv('파일명') | 파일명의 데이터를 불러온다. | 

> index = False 이면 index('x, y, x')는 파일에 저장되지 않는다. 
>
> 이 때, 저장된 파일을 불러오면 새로운 index(0, 1, 2)가 생성된다.
